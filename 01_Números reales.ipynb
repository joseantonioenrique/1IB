## 1.5 Errores.

Las ciencias se encuentra inevitablemente
sujeto a error, y por tanto el tema de cuantificación de errores imprescindible. Hay dos maneras clásicas de medir errores, mediante el **error absoluto** y el **error relativo**. Para definir lo anterior, suponer que $\overline{x}$ es una aproximación al valor
verdadero $x$. El error absoluto es $ε = |x -\overline{x}|$, mientras que el error relativo es $ε_r = \dfrac{|x -\overline{x}|}{|x|}$.
con  $x\neq 0$, donde es necesario resaltar que el error absoluto puede ser algunas veces engañoso, pues no tiene en consideración el tamaño del valor, como si es el caso del error relativo. <br>
No siempre es posible tener el valor verdadero, iremos encontrando
aproximaciones que en cada paso sean mejores, y por tanto, la distancia entre una aproximación y la anterior sea cada vez menor. Esto se conoce como el error aproximado.


###  **Las cifras significativas**
de un número son aquellos dígitos que llevan información significativa sobre la precisión del número. Esto incluye todos los dígitos excepto:
    * Los ceros iniciales que preceden los dígitos no cero.
    * Los ceros finales en un número entero sin un punto decimal explícito

1002 tiene cuatro cifras significativas, pero 0.0032 solo tiene dos. 0.0701 tiene tres cifras significativas (7, 0, y 1).





````python
import math  # son un conjunto de funciones ya programas

def redondear_cifras_significativas(numero, cifras_significativas):
    if numero == 0:
        return 0
    else:
        # Calcular la escala, math.floor redondea al entero menor más cercano, math.ceil() lo hace al mas alto
        #math.ceil(4.3) da 5
        #math.floor(4.3) nos da 4
        # math.log10 hace el logaritmo en base 10 del número.
        escala = math.floor(math.log10(abs(numero))) + 1  # con math.log10 nos da el número de cifras que tiene el número antes de la coma.
        print(escala)
        factor = 10**(cifras_significativas - escala) # para mover el decimal a la posición deseada donde se quiere redondear
        return round(numero * factor) / factor # movemos la coma, multiplicando, redondeamos, y luego volvemos a mover la coma donde estaba.


numero = 12345.99978955678
cifras = 6
resultado = redondear_cifras_significativas(numero, cifras)
print(f"El número {numero} redondeado a {cifras} cifras significativas es: {resultado}")
#observa que no es lo mismo la función round
print(round(numero,6))  # round no es lo mismo, toma 6 difras decimales, la función anterior, toma 6 cifras significativas. ¿entiendes la diferencia?.
````

###  **Error por truncamiento**

El error por truncamiento ocurre cuando se interrumpe un cálculo antes de alcanzar la precisión deseada, utilizando un número finito de términos de una serie o una cantidad limitada de dígitos significativos. Este error es común en cálculos numéricos, como la estimación de raíces cuadradas. Veamos un ejemplo con $\sqrt{2}$
Vamos a usar el método de Newton-Raphson. Es un método por reiteración que usamos para determinar la raíz cuadrada de cualquier número.

$$\sqrt{x}=1+\dfrac{x-1}{2}-\dfrac{(x-1)^2}{8}+\dfrac{(x-1)^3}{16}+\cdots$$
aplicado a $x=2$ tenemos
$$\sqrt{2}=1+\dfrac{2-1}{2}-\dfrac{(2-1)^2}{8}+\dfrac{(2-1)^3}{16}+\cdots$$
es decir
$$2\approx 1+\dfrac{1}{2}-\dfrac{1}{8}+\dfrac{1}{16}​$$
$$2\approx 1+0.5-0.125+0.06252$$
$2≈1.43752​≈1.4375$

Si lo comparamos con el valor real tenemos
El valor exacto de $\sqrt{2}$​ es aproximadamente $1.414213562373095$.
Error por truncamiento=$|Valor_{exacto}-Valor_{truncado}|=|1.414213562373095-1.4375|≈0.023286437626905$
Ahora en Python


````python
import math

# Valor exacto de sqrt(2)
valor_exacto = math.sqrt(2)

# Aproximación truncada
aproximacion_truncada = 1 + 1/2 - 1/8 + 1/16

# Cálculo del error por truncamiento
error_truncamiento = abs(valor_exacto - aproximacion_truncada)

print(f"Valor exacto de sqrt(2): {valor_exacto}")
print(f"Aproximación truncada de sqrt(2): {aproximacion_truncada}")
print(f"Error por truncamiento: {error_truncamiento}")
print(f"Error relativo: {round((error_truncamiento / valor_exacto)*100)} %")
````

###Error por redondeo
Todos conocemos como redondeamos un número según las cifras decimales, que no significativas, que queremos tomar. Veamos un ejemplo en Python con el número $\pi$.


````python
import math

# Valor exacto de pi
valor_exacto = math.pi

# Aproximación redondeada de pi a 4 cifras decimales
aproximacion_redondeada = round(valor_exacto, 4)

# Cálculo del error por redondeo
error_redondeo = abs(valor_exacto - aproximacion_redondeada)

print(f"Valor exacto de pi: {valor_exacto}")
print(f"Aproximación redondeada de pi a 4 cifras decimales: {aproximacion_redondeada}")
print(f"Error relativo: {(error_redondeo/valor_exacto)}")
````
